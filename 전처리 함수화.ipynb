```python
#카테고리 픽스
def category_fix(df):
    #하위 카테고리 픽스
    df.loc[(df['ct_cl'] == '스포츠') & (df['genre_of_ct_cl'] == '레슬링'), 'category_l'] = '스포츠'
    df.loc[(df['ct_cl'] == '스포츠') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '스포츠'

    df.loc[(df['ct_cl'] == '키즈') & (df['genre_of_ct_cl'] == '오락'), 'category_l'] = '키즈'
    df.loc[(df['ct_cl'] == '키즈') & (df['genre_of_ct_cl'] == '학습'), 'category_l'] = '키즈'
    df.loc[(df['ct_cl'] == '키즈') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '키즈'
    df.loc[(df['ct_cl'] == '키즈') & (df['genre_of_ct_cl'] == '애니메이션'), 'category_l'] = '키즈'

    df.loc[(df['ct_cl'] == '공연/음악') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '예능'

    df.loc[(df['ct_cl'] == 'TV드라마') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '기타'
    df.loc[(df['ct_cl'] == 'TV드라마') & (df['genre_of_ct_cl'] == '외화 시리즈'), 'category_l'] = '외화시리즈'

    df.loc[(df['ct_cl'] == '다큐') & (df['genre_of_ct_cl'] == '자연'), 'category_l'] = '교양/다큐멘터리'
    df.loc[(df['ct_cl'] == '다큐') & (df['genre_of_ct_cl'] == '인물'), 'category_l'] = '교양/다큐멘터리'
    df.loc[(df['ct_cl'] == '다큐') & (df['genre_of_ct_cl'] == '역사'), 'category_l'] = '교양/다큐멘터리'

    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '명랑/코믹'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '무협/환타지'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '추리/미스터리'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '액션/모험'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '학원/순정/연애'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '스포츠'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == 'SF/메카'), 'category_l'] = '애니메이션'

    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '추리/미스터리'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '추리/미스터리'), 'category_l'] = '애니메이션'

    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '멜로'), 'category_l'] = '멜로'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '추리/드라마'), 'category_l'] = '추리/드라마'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '드라마'), 'category_l'] = '드라마'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '역사'), 'category_l'] = '사극'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '무협'), 'category_l'] = '무협'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '코미디'), 'category_l'] = '코미디'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '액션/어드벤쳐'), 'category_l'] = '액션/어드벤쳐'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '애니메이션'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '로맨틱코미디'), 'category_l'] = '로맨틱코미디'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == 'SF/환타지'), 'category_l'] = 'SF/환타지'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '다큐멘터리'), 'category_l'] = '다큐멘터리'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '공포/스릴러'), 'category_l'] = '공포/스릴러'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '단편'), 'category_l'] = '단편'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '뮤지컬'), 'category_l'] = '뮤지컬'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '서부'), 'category_l'] = '서부'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '기타'

    df.loc[(df['ct_cl'] == '우리동네') & (df['genre_of_ct_cl'] == '연예/오락'), 'category_l'] = '예능'
    df.loc[(df['ct_cl'] == '우리동네') & (df['genre_of_ct_cl'] == '뉴스'), 'category_l'] = '뉴스'

    df.loc[(df['ct_cl'] == 'TV 연예/오락') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '예능'

    df.loc[(df['ct_cl'] == 'TV 시사/교양') & (df['genre_of_ct_cl'] == '인물/다큐'), 'category_l'] = '교양/다큐멘터리'
    df.loc[(df['ct_cl'] == 'TV 시사/교양') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '교양/다큐멘터리'

    df.loc[(df['ct_cl'] == '라이프') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '예능'
    df.loc[(df['ct_cl'] == '라이프') & (df['genre_of_ct_cl'] == '운동/건강'), 'category_l'] = '예능'

    #상위 카테고리 픽스
    df.loc[df['ct_cl'] == '스포츠', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == '키즈', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == '공연/음악', 'category_h'] = 'TV방송'

    df.loc[df['ct_cl'] == 'TV드라마', 'category_h'] = 'TV드라마'

    df.loc[df['ct_cl'] == '다큐', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == 'TV애니메이션', 'category_h'] = 'TV방송'

    df.loc[df['ct_cl'] == '영화', 'category_h'] = '영화'

    df.loc[df['ct_cl'] == '우리동네', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == 'TV 연예/오락', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == 'TV 시사/교양', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == '라이프', 'category_h'] = 'TV방송'


    # 'category_I' 값이 '기타'이거나 '외화시리즈'이면 category_h의 값을 넣음
    mask1 = df['category_l'] == '기타'
    mask2 = df['category_l'] == '외화시리즈'
    df.loc[mask1, 'category_l'] = df.loc[mask1, 'category_h']
    df.loc[mask2, 'category_l'] = df.loc[mask2, 'category_h']

    return df

# 정규 표현식을 사용하여 괄호와 괄호 안에 있는 문자열을 지우는 함수
def remove_brackets(text):
    return re.sub(r'\([^)]*\)', '', text)

# 시:분:초를 초로 변환하는 함수
def convert_to_sec(time_val):
    return time_val.hour * 3600 + time_val.minute * 60 + time_val.second

def vod_preprocessing(vod):
    # 'asset_nm' 열의 데이터 유형을 문자열로 변환
    vod['asset_nm'] = vod['asset_nm'].astype(str)

    # (예고) 또는 (예고편)이 포함된 행 제거
    vod = vod[~vod['asset_nm'].str.contains('(예고)|(예고편)')]

    # 'disp_rtm'을 'HH:MM:SS' 형식으로 포맷
    vod['disp_rtm'] = pd.to_datetime(vod['disp_rtm'], format='%H:%M')

    # 데이터프레임의 disp_rtm 열을 초로 변환
    vod['disp_rtm'] = vod['disp_rtm'].apply(convert_to_sec)

    #같은 사용자가 같은 컨텐츠를 여러번에 걸쳐서 시청한 경우도 있으므로 subsr이 같고 asset_nm이 같은 경우 use_tms 합쳐버리기
    vod = vod.groupby(['subsr', 'asset_nm']).agg({
    'ct_cl': 'first',
    'genre_of_ct_cl': 'first',
    'use_tms': 'sum',
    'SMRY': 'first',
    'ACTR_DISP': 'first',
    'disp_rtm' : 'first',
    'strt_dt': 'first',
    }).reset_index()

    # use_tms 컬럼이 0이거나 dis_rtm 컬럼이 0인 행 제거
    vod = vod[(vod['use_tms'] != 0) | (vod['disp_rtm'] != 0)]

    # use_tms와 strt_dt_sec를 기반으로 시청률 파생변수 생성
    vod['view_rate'] = (vod['use_tms'] / vod['disp_rtm']) * 100

    # view_rate가 100.000000보다 크면 100.000000으로 변경
    vod.loc[vod['view_rate'] > 100.000000, 'view_rate'] = 100.000000

    # SMRY 컬럼 이상 문장 제거
    words_to_remove = ['서비스', '해지', '시청', '회차', '콘텐츠', '해설', '자막']
    vod['SMRY'] = vod['SMRY'].apply(lambda x: '.'.join([sentence for sentence in str(x).split('.') if not any(word in sentence for word in words_to_remove)]))
    summary_preproc = vod['SMRY'].unique()
    # 파일을 쓰기 모드로 열고 내용을 파일에 쓰기
    with open('summary_preproc.txt', 'w') as file:
        for item in summary_preproc:
            file.write("%s\n\n\n" % item)

    #asset_nm ()안 데이터 제거
    vod['asset_nm'] = vod['asset_nm'].apply(remove_brackets)

    #카테고리 픽스
    # 'genre_of_ct_cl'이 '기타'이면 ct_cl의 값을 넣음
    #mask = vod['genre_of_ct_cl'] == '기타'
    #vod.loc[mask, 'genre_of_ct_cl'] = vod.loc[mask, 'ct_cl']

    # log_dt 칼럼을 날짜 형식으로 변환
    vod['strt_dt'] = pd.to_datetime(vod['strt_dt'], format='%Y%m%d%H%M%S')

    #날짜 정보
    vod['date'] = vod['strt_dt'].dt.strftime('%Y-%m-%d')

    # time 열을 시:분:초 형식으로 합치기
    vod['time'] = vod['strt_dt'].dt.strftime('%H:%M:%S')

    # 요일을 문자열로 변환하여 추가
    vod['weekday'] = vod['strt_dt'].dt.day_name()

    columns_to_drop = ['strt_dt']
    vod = vod.drop(columns=columns_to_drop)

    return vod


def two_month_vod(vod1, vod2):
    vod_12 = pd.concat([vod1, vod2], ignore_index=True)
    vod_pp_df = vod_preprocessing(vod_12)
    vod_df = category_fix(vod_pp_df)

    return vod_df
```

```python
def category_fix(df):
    #하위 카테고리 픽스
    df.loc[(df['ct_cl'] == '스포츠') & (df['genre_of_ct_cl'] == '레슬링'), 'category_l'] = '스포츠'
    df.loc[(df['ct_cl'] == '스포츠') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '스포츠'

    df.loc[(df['ct_cl'] == '키즈') & (df['genre_of_ct_cl'] == '오락'), 'category_l'] = '키즈'
    df.loc[(df['ct_cl'] == '키즈') & (df['genre_of_ct_cl'] == '학습'), 'category_l'] = '키즈'
    df.loc[(df['ct_cl'] == '키즈') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '키즈'
    df.loc[(df['ct_cl'] == '키즈') & (df['genre_of_ct_cl'] == '애니메이션'), 'category_l'] = '키즈'

    df.loc[(df['ct_cl'] == '공연/음악') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '예능'

    df.loc[(df['ct_cl'] == 'TV드라마') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '기타'
    df.loc[(df['ct_cl'] == 'TV드라마') & (df['genre_of_ct_cl'] == '외화 시리즈'), 'category_l'] = '외화시리즈'

    df.loc[(df['ct_cl'] == '다큐') & (df['genre_of_ct_cl'] == '자연'), 'category_l'] = '교양/다큐멘터리'
    df.loc[(df['ct_cl'] == '다큐') & (df['genre_of_ct_cl'] == '인물'), 'category_l'] = '교양/다큐멘터리'
    df.loc[(df['ct_cl'] == '다큐') & (df['genre_of_ct_cl'] == '역사'), 'category_l'] = '교양/다큐멘터리'

    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '명랑/코믹'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '무협/환타지'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '추리/미스터리'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '액션/모험'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '학원/순정/연애'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '스포츠'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == 'SF/메카'), 'category_l'] = '애니메이션'

    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '추리/미스터리'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == 'TV애니메이션') & (df['genre_of_ct_cl'] == '추리/미스터리'), 'category_l'] = '애니메이션'

    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '멜로'), 'category_l'] = '멜로'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '추리/드라마'), 'category_l'] = '추리/드라마'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '드라마'), 'category_l'] = '드라마'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '역사'), 'category_l'] = '사극'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '무협'), 'category_l'] = '무협'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '코미디'), 'category_l'] = '코미디'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '액션/어드벤쳐'), 'category_l'] = '액션/어드벤쳐'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '애니메이션'), 'category_l'] = '애니메이션'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '로맨틱코미디'), 'category_l'] = '로맨틱코미디'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == 'SF/환타지'), 'category_l'] = 'SF/환타지'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '다큐멘터리'), 'category_l'] = '다큐멘터리'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '공포/스릴러'), 'category_l'] = '공포/스릴러'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '단편'), 'category_l'] = '단편'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '뮤지컬'), 'category_l'] = '뮤지컬'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '서부'), 'category_l'] = '서부'
    df.loc[(df['ct_cl'] == '영화') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '기타'

    df.loc[(df['ct_cl'] == '우리동네') & (df['genre_of_ct_cl'] == '연예/오락'), 'category_l'] = '예능'
    df.loc[(df['ct_cl'] == '우리동네') & (df['genre_of_ct_cl'] == '뉴스'), 'category_l'] = '뉴스'

    df.loc[(df['ct_cl'] == 'TV 연예/오락') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '예능'

    df.loc[(df['ct_cl'] == 'TV 시사/교양') & (df['genre_of_ct_cl'] == '인물/다큐'), 'category_l'] = '교양/다큐멘터리'
    df.loc[(df['ct_cl'] == 'TV 시사/교양') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '교양/다큐멘터리'

    df.loc[(df['ct_cl'] == '라이프') & (df['genre_of_ct_cl'] == '기타'), 'category_l'] = '예능'
    df.loc[(df['ct_cl'] == '라이프') & (df['genre_of_ct_cl'] == '운동/건강'), 'category_l'] = '예능'

    #상위 카테고리 픽스
    df.loc[df['ct_cl'] == '스포츠', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == '키즈', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == '공연/음악', 'category_h'] = 'TV방송'

    df.loc[df['ct_cl'] == 'TV드라마', 'category_h'] = 'TV드라마'

    df.loc[df['ct_cl'] == '다큐', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == 'TV애니메이션', 'category_h'] = 'TV방송'

    df.loc[df['ct_cl'] == '영화', 'category_h'] = '영화'

    df.loc[df['ct_cl'] == '우리동네', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == 'TV 연예/오락', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == 'TV 시사/교양', 'category_h'] = 'TV방송'
    df.loc[df['ct_cl'] == '라이프', 'category_h'] = 'TV방송'


    # 'category_I' 값이 '기타'이거나 '외화시리즈'이면 category_h의 값을 넣음
    mask1 = df['category_l'] == '기타'
    mask2 = df['category_l'] == '외화시리즈'
    df.loc[mask1, 'category_l'] = df.loc[mask1, 'category_h']
    df.loc[mask2, 'category_l'] = df.loc[mask2, 'category_h']

    return df

# 정규 표현식을 사용하여 괄호와 괄호 안에 있는 문자열을 지우는 함수
def remove_brackets(text):
    return re.sub(r'\([^)]*\)', '', text)

# 시:분:초를 초로 변환하는 함수
def convert_to_sec(time_val):
    return time_val.hour * 3600 + time_val.minute * 60 + time_val.second

def contents_preprocessing(contents):
    # contents_asset에서 super_asset_nm 컬럼명을 asset_nm으로 변경
    contents = contents.rename(columns={'super_asset_nm': 'asset_nm'})

    # 'asset_nm' 열의 데이터 유형을 문자열로 변환
    contents['asset_nm'] = contents['asset_nm'].astype(str)

    # (예고) 또는 (예고편)이 포함된 행 제거
    contents = contents[~contents['asset_nm'].str.contains('(예고)|(예고편)')]

    # 'disp_rtm'을 'HH:MM:SS' 형식으로 포맷
    contents['disp_rtm'] = pd.to_datetime(contents['disp_rtm'], format='%H:%M')

    # 데이터프레임의 disp_rtm 열을 초로 변환
    contents['disp_rtm'] = contents['disp_rtm'].apply(convert_to_sec)

    # disp_rtm 컬럼이 0인 행 제거
    contents = contents[contents['disp_rtm'] != 0]

    # SMRY 컬럼 이상 문장 제거
    words_to_remove = ['서비스', '해지', '시청', '회차', '콘텐츠', '해설', '자막']
    contents['SMRY'] = contents['SMRY'].apply(lambda x: '.'.join([sentence for sentence in str(x).split('.') if not any(word in sentence for word in words_to_remove)]))
    summary_preproc = contents['SMRY'].unique()

    # 파일을 쓰기 모드로 열고 내용을 파일에 쓰기
    with open('summary_preproc.txt', 'w') as file:
        for item in summary_preproc:
            file.write("%s\n\n\n" % item)

    # asset_nm ()안 데이터 제거
    contents['asset_nm'] = contents['asset_nm'].apply(remove_brackets)

    # 카테고리 픽스

    # 'genre_of_ct_cl'이 '기타'이면 ct_cl의 값을 넣음
    #mask = contents['genre_of_ct_cl'] == '기타'
    #contents.loc[mask, 'genre_of_ct_cl'] = contents.loc[mask, 'ct_cl']

    # log_dt 칼럼을 날짜 형식으로 변환
    contents['log_dt'] = pd.to_datetime(contents['log_dt'], format='%Y%m%d%H%M%S')  # 수정된 부분

    #날짜 정보
    contents['date'] = contents['log_dt'].dt.strftime('%Y-%m-%d')

    # time 열을 시:분:초 형식으로 합치기
    contents['time'] = contents['log_dt'].dt.strftime('%H:%M:%S')

    # 요일을 문자열로 변환하여 추가
    contents['weekday'] = contents['log_dt'].dt.day_name()

    columns_to_drop = ['log_dt']
    contents = contents.drop(columns=columns_to_drop)

    return contents

def two_month_contents(contents1, contents2):
    contents_12 = pd.concat([contents1, contents2], ignore_index=True)
    contents_pp_df = contents_preprocessing(contents_12)
    contents_df = category_fix(contents_pp_df)

    return contents_df
```

```python
def convert_to_hour(time):
    return pd.to_datetime(time, format='%H:%M:%S').strftime('%H')

def time_view_df(vod_df):

    # 'time' 열을 Datetime으로 변환
    vod_df['time'] = pd.to_datetime(vod_df['time'])

    start_time = pd.to_datetime('00:00', format='%H:%M')
    end_time = pd.to_datetime('23:59:59', format='%H:%M:%S')  # 범위를 23:59:59로 확장
    time_range = pd.date_range(start=start_time, end=end_time, freq='1H')  # Change frequency to 1 hour
    grouped_vod_df = []

    for i in range(len(time_range)):
        start = time_range[i].time()

        # 현재 범위의 종료 시간을 결정
        if i == len(time_range) - 1:
            # 만약 마지막 범위라면 종료 시간을 23:59:59로 설정
            end = pd.to_datetime('23:59:59', format='%H:%M:%S').time()
        else:
            end = time_range[i + 1].time()

        start_datetime = pd.to_datetime(start.strftime('%H:%M'), format='%H:%M')
        end_datetime = pd.to_datetime(end.strftime('%H:%M'), format='%H:%M')

        filtered_df = vod_df[(vod_df['time'].dt.time >= start_datetime.time()) & (vod_df['time'].dt.time <= end_datetime.time())]

        if not filtered_df.empty:
            genre_counts = filtered_df['category_l'].value_counts().nlargest(5)
            top_genres = ", ".join(genre_counts.index.astype(str).tolist())
        else:
            top_genres = ""

        if not filtered_df.empty:
            asset_nm_counts = filtered_df['asset_nm'].value_counts().nlargest(5)
            top_asset_nm = ", ".join(map(str, asset_nm_counts.index.tolist()))
        else:
            top_asset_nm = ""

        temp_df = pd.DataFrame({"time_range": start, "top_genre": [top_genres], "top_asset": [top_asset_nm]})
        grouped_vod_df.append(temp_df)

    result_df = pd.concat(grouped_vod_df, ignore_index=True)
    # time_range 컬럼을 새로운 형식으로 변환
    result_df['time_range'] = result_df['time_range'].apply(convert_to_hour)

    return result_df
```

```python
def weekday(vod_df):

    time_range = 24
    grouped_vod_df = []


    filtered_df = vod_df.loc[(vod_df['weekday'] != 'Saturday') & (vod_df['weekday'] != 'Sunday')]

    if not filtered_df.empty:
        genre_counts = filtered_df['category_l'].value_counts().nlargest(5)
        top_genres = ", ".join(genre_counts.index.astype(str).tolist())
    else:
        top_genres = ""

    if not filtered_df.empty:
        asset_nm_counts = filtered_df['asset_nm'].value_counts().nlargest(5)
        top_asset_nm = ", ".join(map(str, asset_nm_counts.index.tolist()))
    else:
        top_asset_nm = ""

    temp_df = pd.DataFrame({"time_range": time_range, "top_genre": [top_genres], "top_asset": [top_asset_nm]})
    grouped_vod_df.append(temp_df)

    weekday_df = pd.concat(grouped_vod_df, ignore_index=True)

    return weekday_df
```

```python
def friday(vod_df):

    time_range = 26
    grouped_vod_df = []


    filtered_df = vod_df.loc[vod_df['weekday'] == 'Friday']

    if not filtered_df.empty:
        genre_counts = filtered_df['category_l'].value_counts().nlargest(5)
        top_genres = ", ".join(genre_counts.index.astype(str).tolist())
    else:
        top_genres = ""

    if not filtered_df.empty:
        asset_nm_counts = filtered_df['asset_nm'].value_counts().nlargest(5)
        top_asset_nm = ", ".join(map(str, asset_nm_counts.index.tolist()))
    else:
        top_asset_nm = ""

    temp_df = pd.DataFrame({"time_range": time_range, "top_genre": [top_genres], "top_asset": [top_asset_nm]})
    grouped_vod_df.append(temp_df)

    friday_df = pd.concat(grouped_vod_df, ignore_index=True)

    return friday_df
```

```python
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

def remove_brackets_and_square_brackets(text):
    # 괄호 () 제거
    text = re.sub(r'\([^)]*\)', '', text)
    # 대괄호 [] 제거
    text = re.sub(r'\[[^\]]*\]', '', text)
    return text

def preprocess_summary_column(asset_df, words_to_remove=['서비스', '해지', '시청', '회차', '콘텐츠', '해설', '자막']):
    asset_df['SMRY'] = asset_df['SMRY'].apply(lambda x: '.'.join([sentence for sentence in str(x).split('.') if not any(word in sentence for word in words_to_remove)]))
    
    with open('summary_preproc.txt', 'w') as file:
        for item in asset_df['SMRY'].unique():
            file.write("%s\n\n\n" % item)

    asset_df['SMRY'] = asset_df['SMRY'].apply(remove_brackets_and_square_brackets)
    asset_df = asset_df[asset_df['SMRY'] != '']
    return asset_df

def find_similar_assets(asset_df, n_neighbors=11):
    asset_df['SMRY'] = asset_df['SMRY'].fillna('')

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(asset_df['SMRY'])

    knn = NearestNeighbors(n_neighbors=n_neighbors, metric='cosine')
    knn.fit(tfidf_matrix)

    asset_df.reset_index(drop=True, inplace=True)

    similar_df = pd.DataFrame(columns=['asset_nm', 'similar_assets'])

    for i, row in asset_df.iterrows():
        _, indices = knn.kneighbors(tfidf_matrix[i])
        # 자기 자신을 제외하고 10개까지만 선택
        similar_assets = asset_df['asset_nm'].iloc[indices[0][1:]].tolist()[:n_neighbors-1]
        similar_df = similar_df.append({'asset_nm': row['asset_nm'], 'similar_assets': similar_assets}, ignore_index=True)

    return similar_df
```

```python
reco4 = find_similar_assets(asset_df)
```

```python
reco4.to_csv(r'reco4.csv', index=False, encoding='cp949', errors='replace')
```
