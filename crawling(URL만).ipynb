```python
import pandas as pd

# 'asset_nm' 열만 추출하여 새로운 데이터프레임 생성
vod_asset_df = vod_df[['asset_nm']].copy()

# 'super_asset_nm' 열만 추출하여 새로운 데이터프레임 생성 및 괄호 및 괄호 안 문자열 제거
contents_asset_df = contents_df[['super_asset_nm']].copy()
contents_asset_df['super_asset_nm'] = contents_asset_df['super_asset_nm'].str.replace(r'\([^)]*\)', '').str.strip()

# 'super_asset_nm' 컬럼을 'asset_nm'으로 변경
contents_asset_df.rename(columns={'super_asset_nm': 'asset_nm'}, inplace=True)

# vod_asset_df와 contents_asset_df 데이터프레임 수직으로 병합
total_df = pd.concat([vod_asset_df, contents_asset_df], ignore_index=True)

print(len(total_df))
```

```python
total_df['asset_nm'] = total_df['asset_nm'].astype(str)

# 정규 표현식을 사용하여 괄호와 괄호 안에 있는 문자열을 지우는 함수
def remove_brackets(text):
    return re.sub(r'\([^)]*\)', '', text)

# apply 함수를 사용하여 데이터프레임의 asset_nm 컬럼에 함수를 적용합니다.
total_df['asset_nm'] = total_df['asset_nm'].apply(remove_brackets)

#회차 정보 제거
for idx, row in total_df.iterrows():
    title = row['asset_nm']
    processed_title = re.sub(r'\s[0-9]+회', '', title)  # 정규 표현식을 사용하여 '숫자회' 부분을 공백으로 대체합니다.
    total_df.at[idx, 'asset_nm'] = processed_title  # 수정된 드라마 제목을 데이터프레임에 다시 할당합니다.

# 중복된 항목 제거
total_df = total_df.drop_duplicates()

    
# 수정된 데이터프레임 확인
total_df.head()
```

```python
def crawl(df):
    # webdriver를 시작합니다.
    driver = webdriver.Chrome()  # 또는 원하는 다른 브라우저를 선택할 수 있습니다.

    url = "https://www.themoviedb.org/"

    # 각 데이터를 저장할 빈 리스트 생성
    nm_lst = []
    img_lst = []
    missing_lst = []

    # 데이터프레임의 asset_nm 컬럼에서 영화 제목을 가져와 크롤링을 수행합니다.
    length = len(df)

    for index, row in df.iterrows():
        driver.implicitly_wait(3)
        driver.get(url)
        search_input = driver.find_element(By.XPATH, '/html/body/div[1]/main/section[1]/div/div/div/div[2]/form/label/input')
        search_input.send_keys(row['asset_nm'])
        search_input.send_keys(Keys.RETURN)

        try:
            search_button = WebDriverWait(driver, 3).until(EC.visibility_of_element_located((By.XPATH, '//*[contains(@id, "card_tv_")]/div/div[2]/div[1]/div/div/a/h2')))
            search_button.click()
        except TimeoutException:
            try:
                search_button = WebDriverWait(driver, 3).until(EC.visibility_of_element_located((By.XPATH, '//*[contains(@id, "card_movie_")]/div/div[2]/div[1]/div/div/a/h2')))
                search_button.click()
            except TimeoutException:
                print(f"요소가 시간 내에 로드되지 않았습니다: {row['asset_nm']}")
                missing_lst.append(row['asset_nm'])
                continue

        # 이름 nm
        try:
            nm = driver.find_element(By.XPATH, '//*[@id="original_header"]/div[2]/section/div[1]/h2/a').text
        except NoSuchElementException:
            nm = 'none'
            
        #이미지 img
        try:
            img = driver.find_element(By.XPATH, '//*[@id="original_header"]/div[1]/div/div[1]/img').get_attribute('src')
        except NoSuchElementException:
            img = 'none'


        # 각 리스트에 저장
        nm_lst.append(nm)
        img_lst.append(img)

    # pandas DataFrame을 직접 만들어 리스트를 컬럼에 할당합니다.
    crawl_df = pd.DataFrame({
        'name': nm_lst,
        'image': img_lst
    })
    # missing_lst 리스트를 DataFrame으로 변환
    missing_df = pd.DataFrame({'missing_column_name': missing_lst})

    #  ★★★개인 노트북에서 하실 때 파일 path 꼭 변경 해 주셔야 합니다★★★
    crawl_df.to_csv(r'C:\Users\hanji\크롤링결과\crawl_df.csv', index=False, encoding='cp949', errors='replace')
    missing_df.to_csv(r'C:\Users\hanji\크롤링결과\missing_df.csv', index=False, encoding='cp949', errors='replace')
    return crawl_df
```
